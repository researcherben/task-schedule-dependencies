```python
G = pygraphviz.AGraph(directed=True, compound=True)
G.add_node("instance_123_persistent_44", 
           label="start here")

G.add_node("instance_492_persistent_49", 
           label="description: next task\l"+
                 "this task (cost,dur)=[(1,2), (3,4)]\l"+
                 "cumulative (cost,dur)=[(1, 2), (3, 4)]\l"+
                 "output param=['a', 'b']\l",
           shape="rect");

G.add_edge("instance_123_persistent_44","instance_492_persistent_49")


G.add_node("instance_442_persistent_59", 
           label="if a>5:\l"+
           "description: another task\l"+
           "this task (cost,dur)=[(11,2), (9,4)]\l"+
           "cumulative (cost,dur)=[(1, 2), (3, 4)]\l"+
           "output param=['e', 'f']\l",
           shape="rect")
 
G.add_node("instance_412_persistent_52", 
           label="if a<=5:\l"+
           "description: some task\l"+
           "this task (cost,dur)=[(10,2), (9,4)]\l"+
           "cumulative (cost,dur)=[(1, 2), (3, 4)]\l"+
           "output param=['c', 'd']\l",
           shape="rect")
 
G.add_edge("instance_492_persistent_49","instance_412_persistent_52")
 
G.add_edge("instance_492_persistent_49","instance_442_persistent_59")
 
G.add_node("instance_22_persistent_23", 
           label="if d>=4:\l"+
           "description: la task\l"+
           "this task (cost,dur)=[(6,2), (9,4)]\l"+
           "cumulative (cost,dur)=[(1, 2), (3, 4)]\l"+
           "output param=['c', 'd']\l",
           shape="rect");
 
G.add_edge("instance_412_persistent_52","instance_22_persistent_23")

# ********** begin subgraph ***************
# https://pygraphviz.github.io/documentation/stable/auto_examples/plot_subgraph.html
B = G.add_subgraph(name="cluster_instance_99_persistent_41",
                   label = "if d<4:"+
   "description: a task\l"+
   "this task (cost,dur)=[(4,2), (9,4)]\l"+
   "cumulative (cost,dur)=[(1, 2), (3, 4)]\l"+
   "output param=['l', 'm']\l"
)

B.add_node("instance_99_persistent_41", style="invis");

B.add_node("instance_12_persistent_33", 
           label="description: the task\l"+
                 "this task (cost,dur)=[(3,2), (9,4)]\l"+
                 "cumulative (cost,dur)=[(1, 2), (3, 4)]\l"+
                 "output param=['k', 'p']\l",
           shape="rect")

B.add_node("instance_11_persistent_13",
           label="description: yet another task\l"+
                 "this task (cost,dur)=[(2,2), (9,4)]\l"+
                 "cumulative (cost,dur)=[(1, 2), (3, 4)]\l"+
                 "output param=['l', 'm']\l",
           shape="rect")
  
B.add_node("instance_10_persistent_03",
           label="description: le task\l"+
                 "this task (cost,dur)=[(1,2), (9,4)]\l"+
                 "cumulative (cost,dur)=[(1, 2), (3, 4)]\l"+
                 "output param=['l', 'm']",
           shape="rect")
  
B.add_edge("instance_12_persistent_33","instance_11_persistent_13")
B.add_edge("instance_12_persistent_33","instance_10_persistent_03")

# ********** end subgraph ***************

G.add_edge("instance_412_persistent_52","instance_99_persistent_41", 
           lhead="cluster_instance_99_persistent_41")
 
G.add_node("instance_111_persistent_22",
           label="description: later task\l"+
                 "this task (cost,dur)=[(1,2), (3,4)]\l"+
                 "cumulative (cost,dur)=[(1, 2), (3, 4)]\l"+
                 "output param=['q', 'r']\l",
           shape="rect")

G.add_edge("instance_99_persistent_41","instance_111_persistent_22", 
           ltail="cluster_instance_99_persistent_41")
```
